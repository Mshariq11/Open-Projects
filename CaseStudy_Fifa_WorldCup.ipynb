## _**DATA SOURCE AND BRIEF DESCRITPTION:**_

The data for this analysis is sourced from Kaggle. The dataset includes information on matches from various World Cups, featuring attributes such as:

- YEAR: The year the match took place.
- DATETIME: The specific date and time of the match.
- STAGE: The tournament stage (e.g., group, knockout).
- STADIUM: The venue where the match was held.
- CITY: The city hosting the match.
- HOME TEAM NAME: The name of the team playing at home.
- HOME TEAM GOALS: The number of goals scored by the home team.
- AWAY TEAM NAME: The name of the team playing away.
- WIN CONDITIONS: Conditions under which the match was won.
- ATTENDANCE: The number of spectators at the match.
- HALF TIME HOME GOALS: Goals scored by the home team at halftime.
- HALF TIME AWAY GOALS: Goals scored by the away team at halftime.
- REFEREE: The name of the match referee.
- ASSISTANT 1: The first assistant referee's name.
- ASSISTANT 2: The second assistant referee's name.
- ROUND ID: Identifier for the match round.
- MATCH ID: Unique identifier for the match.
- HOME TEAM INITIALS: Abbreviation of the home team's name.
- AWAY TEAM INITIALS: Abbreviation of the away team's name.

```python
# Match outcomes by home and away teams

df_WC_M.head(10)
```

```python
def Winner(df_WC_M):
    if df_WC_M['Home Team Goals'] > df_WC_M['Away Team Goals']:
        return 'Home Team Won'
    elif df_WC_M['Home Team Goals'] < df_WC_M['Away Team Goals']:
        return 'Away Team Won'
```

```python
df_WC_M['Winner'] = df_WC_M.apply(lambda x:Winner(x), axis=1)
```

```python
df_WC_M['Winner'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Assuming you already have the df_WC_M DataFrame
df = df_WC_M['Winner'].value_counts()

# Define the colors
colors = ['#458B74', '#E3CF57']

# Create the donut chart
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)
ax.pie(df, labels=df.index, colors=colors, autopct='%1.1f%%', pctdistance=0.8, labeldistance=1.1, wedgeprops=dict(width=0.5))

# Add a circle at the center to create a donut chart
centre_circle = plt.Circle((0, 0), 0.5, color='white', fc='white', linewidth=0)
ax.add_artist(centre_circle)

# Add a title and Legend
plt.title('Match Outcome', fontsize=20, fontstyle='italic', loc='center')
plt.legend(df.index, title='Match Outcomes', loc='lower left', bbox_to_anchor=(1, 0))

plt.show()
```
