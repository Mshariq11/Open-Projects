## _**1.1-Data Description:**_
 
> The Nashville housing dataset contains information about residential properties in the Nashville area. The dataset includes the following columns:
- UniqueID: Unique identifier for each property
- ParcelID: Unique identifier for each parcel of land
- LandUse: The type of land use (e.g., residential, commercial, etc.)
- PropertyAddress: The physical address of the property
- SaleDate: The date the property was sold
- SalePrice: The price the property was sold for
- LegalReference: The legal reference number for the property sale
- SoldAsVacant: Indicates whether the property was sold as vacant or not
- OwnerName: The name of the property owner
- OwnerAddress: The address of the property owner
- Acreage: The size of the property in acres
- TaxDistrict: The tax district the property is located in
- LandValue: The value of the land
- BuildingValue: The value of the building
- TotalValue: The total value of the property
- YearBuilt: The year the building was constructed
- Bedrooms: The number of bedrooms in the property
- FullBath: The number of full bathrooms in the property
- HalfBath: The number of half bathrooms in the property


### ***2.0-KEY FINDINGS BASED ON Research Questions:***

_1. Factors Influencing House Prices:_
- Economic Indicators: House prices were positively influenced by local economic growth, including job creation and wage increases.
- Location: Proximity to amenities such as schools, parks, and public transportation significantly impacted prices, with desirable neighborhoods commanding higher prices.
- Property Features: Characteristics such as square footage, number of bedrooms, and modern amenities also played a crucial role in determining property values.

_2. Patterns in Housing Transactions:_
- Transaction Volume: The number of transactions increased annually, indicating a strong demand for housing.
- Seasonal Trends: Transactions typically peaked in spring and summer, aligning with traditional home-buying seasons.
- Buyer Demographics: An influx of millennials and young professionals contributed to the rising number of transactions.

_3. Identifying Undervalued or Overvalued Properties:_
- Price-to-Income Ratios: Certain neighborhoods showed signs of overvaluation based on price-to-income ratios, suggesting potential market corrections.
- Comparative Market Analysis: Properties with amenities but lower sale prices indicated undervaluation compared to similar homes in the area.
- Time on Market: Homes that lingered on the market longer than average were often identified as potentially overvalued.

_4. Analyzing Neighborhood Characteristics:_
- Demographic Shifts: Neighborhoods with increasing populations of young families and professionals experienced faster price appreciation.
- Community Features: Areas with higher walkability scores and access to public services saw greater demand and higher property values.
- Crime Rates: Lower crime rates correlated with increased property values, highlighting the importance of safety in housing decisions.

_5. Forecasting Future Housing Market Trends:_
- Continued Growth: Indicators suggested continued price appreciation in the following years, driven by population growth and economic development.
- Supply Constraints: A limited housing supply relative to demand indicated potential price increases, especially in sought-after neighborhoods.
- Market Diversification: Emerging neighborhoods began attracting investment, suggesting shifts in buyer interest and future growth areas.

## *8.0-Standarized The Data / Data PreProcessing*

- Transforms the datatypes of the relevant columns to ensure they are in the correct format for analysis.

- Removes the unused columns from the dataset.

```python
nashville_housing.columns
```

#### *8.1-Transform the datatypes of some columns*

```python
nashville_housing['Sale Date'] = pd.to_datetime(nashville_housing['Sale Date'])
nashville_housing['Sale Price'] = nashville_housing['Sale Price'].astype(float)
```

#### *8.2-Replace the Space*

```python
# The code nashville_housing.replace(r'^\s*$', np.nan) is used to replace all empty string values (i.e., strings containing only whitespace) in a DataFrame df_housing with the np.nan (Not a Number) value.
# r'^\s*$': This is a regular expression pattern that matches any string that contains only whitespace characters (spaces, tabs, newlines, etc.). 
nashville_housing.replace(r'^\s*$', np.nan)
```

#### *8.3-Remove the unused columns*

```python
nashville_housing.drop(['Unnamed: 0.1', 'Unnamed: 0', 'Parcel ID', 'Suite/ Condo   #', 'Legal Reference', 'Owner Name', 'Address', 'City', 'State',
                'image' ], axis=1, inplace=True)
```

```python
nashville_housing.reset_index(inplace=True, drop=True)
nashville_housing.info()
```
