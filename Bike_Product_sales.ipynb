# ***D) Key Findings:***

    The case study identifies key trends and factors that influence bike sales, and provides recommendations on how to increase ROI. This information can be used by businesses to make informed decisions about their bike sales strategies.

Here are the some key insights gained from the Bike Product Sales Data. 

- Bike product sales were higher in 2014 & 2016 (29,398 units) compared to 2013 & 2015 (24,443 units). The lowest sales were in 2011 & 2012 (2,677 units).
- The highest sales months are June, December, May, and April. The lowest sales months are August, September, and July.
- The USA has the highest bike product sales (39,206 units), followed by Australia (23,936 units). Germany and France have the lowest sales (around 11,000 units).
- The median customer age is 35 years old, with a range from 15-87 years old. The customer age distribution is right-skewed, with more young adults and adults.
- Most customers order 1-2 items, with the most common order quantities being 5-27 items. The average order quantity is approximately 12 items.
- The highest order volumes come from the USA (39,206 items) and Australia (23,936 items), with the lowest from France (10,998 items).
- There were 104 bike rack orders from Canada.
- The majority of orders (62%) are for accessories, followed by bikes (23%) and clothing (15%).
- The most popular subcategory is tires and tubes (33,870 units), followed by bottles and cages (15,876 units). Bike stands have the lowest sales (456 units).
- Road bikes are the most popular type, accounting for 51.7% of sales, followed by mountain bikes (34.1%) and touring bikes (14.2%).
- Male customers (51.6%) slightly outnumber female customers (48.4%).
- There are 8 male customers who have spent more than $500 in revenue.

- Focus marketing efforts on the peak sales months (June, December, May, April) to maximize revenue.

#### *BoxPlot*

What is the importance of Box plot?
1. Box plots are a graphical representation of the distribution of data. They can be used to identify outliers, skewness, and other patterns in the data.
2. The minimum and maximum values of the data are represented by the bottom and top of the box, respectively. The **median value** is represented by the line in the middle of the box. The 25th and 75th percentiles are represented by the edges of the box.


      In the case of **customer age**, the box plot can provide insights into the age range of customers who are interested in the product. Such as the minimum value is 17 and the maximum value is 87, it means that the product is likely to appeal to a wide range of customers.
